```
def nextSqure():
  i=1;
  while True:
    yield i*i
    i=i+1
for num in nextSqure():
  if num>100:
    break
  print(num)
```

```
def get_triangle(G1):
  for n1 in G1.nodes:
    neighbors1=set(G1[n1])
    for n2 in filter(lambda x: x>n1,neighbors1):
      neighbors2=set(G1[n2])
      common=neighbors1 & neighbors2
      for n3 in filter(lambda x:x>n2,common):
        yield n1,n2,n3
G1=nx.complete_graph(6)
for triangle in get_triangle(G1):
 print(triangle)
 
```
